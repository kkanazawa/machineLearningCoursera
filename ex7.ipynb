# first part

```python
def dist(a,b):
    dist = sqrt(sum((a-b)**2,axis=1))
    return dist
```

```python
def findClosestCentroids(X, centroids):
    dlist = []
    k = centroids.size/2
    for i in range(k):
        dlist.append( dist(X,centroids[i,:]) )
        
    D = vstack((dlist[0],dlist[1],dlist[2])).T # how should I write? avoid using list?
    idx = argmin(D,axis=1)
    return idx
```

# new first part

```python
from numpy.linalg import norm
```

```python
centroids = initial_centroids
```

```python
def findClosestCentroids2(X, centroids):
    m,n = X.shape
    k = centroids.size/n

    Xp = X.reshape((m,1,n))
    Cp = centroids.reshape((1,k,n))

    D2 = norm(Xp-Cp, axis=2)
    idx2 = argmin(D2, axis=1)
    return idx2
```

```python
findClosestCentroids2(X,initial_centroids)
```

```python
def computeCentroids(X,idx):

    idx0 = (idx == 0).nonzero()[0]
    idx1 = (idx == 1).nonzero()[0]
    idx2 = (idx == 2).nonzero()[0]
    tmp_centroids = vstack((mean(X[idx0,],axis=0),                            
                            mean(X[idx1,],axis=0),
                            mean(X[idx2,],axis=0)))
    return tmp_centroids
```

```python
X[idx==1].shape, X[idx==2].shape, X[idx==0].shape
```

```python
def computeCentroids2(X,idx,k):

    m,n = X.shape
    centroids = array([])

    for i in range(k):
        C5 = X[idx==i].mean(axis=0)
        centroids = append(centroids,C5,axis=1)
        
    centroids.shape = ((k,n))

    return centroids
```

```python
idx2 = findClosestCentroids2(X,initial_centroids)
```

```python
computeCentroids2(X,idx2,3)
```

```python
centroids = initial_centroids
k = 3

for i in range(10):
    idx_tmp = findClosestCentroids2(X,centroids)
    centroids = computeCentroids2(X,idx_tmp,k)

centroids
```
